```python
# 2. Importa√ß√µes e Configura√ß√£o

import google.generativeai as genai

from IPython.display import Markdown

import time

# --- CONFIGURA√á√ÉO DA CHAVE DE API ---

# Cole sua chave API abaixo (mantenha as aspas)

GOOGLE_API_KEY = "AIzaSyBzocKtUNc-JVUpcOr8xhYarenND3-AQPI"

genai.configure(api_key=GOOGLE_API_KEY)

# --- DEFINI√á√ÉO DO MODELO E CONTEXTO ---

model = genai.GenerativeModel('gemini-2.5-flash')

# Aqui definimos a "Personalidade" e o "Conhecimento do Neg√≥cio"

contexto_negocio = """

Voc√™ √© o assistente virtual da 'TechBurger', uma hamburgueria artesanal futurista.

Card√°pio:

- Burger Bit (R$ 20,00): P√£o, carne, queijo e molho especial.

- Chips de Sil√≠cio (R$ 10,00): Batatas r√∫sticas.

- Shake Java (R$ 15,00): Milkshake de caf√©.

Hor√°rio de funcionamento: Das 18h √†s 23h, de ter√ßa a domingo.

Entregas: Apenas num raio de 5km.

Tom de voz: Amig√°vel, nerd e prestativo.

Regra: Responda apenas perguntas sobre a hamburgueria.

"""

# Inicializa o chat com a instru√ß√£o do sistema

chat = model.start_chat(history=[

    {"role": "user", "parts": f"Aja conforme este contexto: {contexto_negocio}"},

    {"role": "model", "parts": "Entendido! Estou pronto para atender os clientes da TechBurger."}

])

historico_conversa = []

print("üçî Bem-vindo ao Chat da TechBurger! (Voc√™ tem direito a 3 perguntas)")

print("-" * 60)

# --- LOOP DE 3 PERGUNTAS ---

for i in range(1, 4):

    pergunta = input(f"\nPergunta {i}/3: ")

    # Envia a pergunta para o Gemini

    response = chat.send_message(pergunta)

    # Mostra a resposta

    print(f"ü§ñ Bot: {response.text}")

    # Salva para o resumo final

    historico_conversa.append(f"Pergunta: {pergunta}\nResposta: {response.text}")

    time.sleep(1) # Pausa pequena para organiza√ß√£o

# --- GERA√á√ÉO DO RESUMO ---

print("\n" + "="*30)

print("üìù GERANDO RESUMO DO ATENDIMENTO...")

print("="*30)

prompt_resumo = f"""

Com base nas intera√ß√µes abaixo, crie um resumo curto e elegante em t√≥picos

do que o cliente solicitou e quais informa√ß√µes foram passadas:

{historico_conversa}

"""

resumo = model.generate_content(prompt_resumo)

# Exibe o resumo formatado

display(Markdown(resumo.text))
```
